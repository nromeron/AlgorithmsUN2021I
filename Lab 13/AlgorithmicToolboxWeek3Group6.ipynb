## Fractional knapsack

```python
%%writefile knapsack.py
# Uses python3
import sys

def get_optimal_value(capacity, weights, values):
    value = 0.
    items = []
    for i in range(len(weights)):
        items.append([values[i],weights[i]])

    items.sort(key = lambda x: x[0]/x[1])

    for v, w in items[::-1]:
        if capacity > 0:
            f = min(capacity,w)
            value += f/w * v
            capacity -= f

    return value


if __name__ == "__main__":
    data = list(map(int, sys.stdin.read().split()))
    n, capacity = data[0:2]
    values = data[2:(2 * n + 2):2]
    weights = data[3:(2 * n + 2):2]
    opt_value = get_optimal_value(capacity, weights, values)
    print("{:.4f}".format(opt_value))
```

##Car Fueling

```python
d = int(input())
m = int(input())
n = int(input())
stops = [int(i) for i in input().split()]
stops.append(d)

capacity = m

refills = 0

for j in range(n+1):
    actual_distance = stops.pop(0)
    if actual_distance > m:
        refills = -1
        break
    m -= actual_distance
    stops = [k - actual_distance for k in stops]
    if stops and stops[0]>m:
        refills += 1
        m = capacity

print(refills)
```

## Max Advertisement Revenue

```python
n = int(input())
a = [int(i) for i in input().split()]
b = [int(i) for i in input().split()]
a.sort()
b.sort()
maximum = sum([a[i]*b[i] for i in range(n)])
print(maximum)
```

## Collecting Signatures

```python
n = int(input())
segments = []

for i in range(n):
    a, b = [int(j) for j in input().split()]
    segments.append([a,b])

segments.sort(key = lambda x: x[1])

point = 0
intersections = []

while point < n:
    curr = segments[point]
    while point < n-1 and curr[1]>=segments[point+1][0]:
        point += 1
    intersections.append(curr[1])
    point += 1

marks = len(intersections)
print(marks)
for p in range(marks-1):
    print(intersections[p],end=' ')
print(intersections[-1])
```

## Maximum Number of Prizes

```python
n = int(input())

limit = n
nums = []

if n == 1:
    nums.append(1)

for i in range(1, n):
    if limit>2*i:
        nums.append(i)
        limit -= i
    else:
        nums.append(limit)
        break

prizes = len(nums)
print(prizes)
for p in range(prizes-1):
    print(nums[p],end=' ')
print(nums[-1])
```

## Maximum Salary

```python
n = int(input())
Digits = [i for i in input().split()]

answer = []

while Digits:
    max_digit = '0'
    for digit in Digits:
        if int(digit+max_digit)>=int(max_digit+digit):
            max_digit = digit
    answer.append(int(max_digit))
    Digits.remove(max_digit)

for p in range(len(answer)-1):
    print(answer[p],end='')
print(answer[-1])
```
