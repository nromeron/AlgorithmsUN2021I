## Maximum amount of Gold


```python
capacity, n = [int(i) for i in input().split()]
weights = [int(i) for i in input().split()]

MaxWeights = [[0]*(n+1) for fila in range(capacity+1)]
for i in range(1, capacity+1):
    for j in range(1, n+1):
        MaxWeights[i][j] = MaxWeights[i][j-1]
        if weights[j-1]<=i:
            opt = MaxWeights[i-weights[j-1]][j-1] + weights[j-1]
            if opt > MaxWeights[i][j]:
                MaxWeights[i][j] = opt

print(MaxWeights[capacity][n])
```

##Partitioning souvenirs

```python
def souvenirs(capacity,n,weights):
    partitions = 0
    MaxWeights = [[0]*(n+1) for fila in range(capacity+1)]
    for i in range(1, capacity+1):
        for j in range(1, n+1):
            MaxWeights[i][j] = MaxWeights[i][j-1]
            if weights[j-1]<=i:
                opt = MaxWeights[i-weights[j-1]][j-1] + weights[j-1]
                if opt > MaxWeights[i][j]:
                    MaxWeights[i][j] = opt
            if MaxWeights[i][j] == capacity:
                partitions += 1
    if partitions < 3:
        return 0
    else:
        return 1

n = int(input())
v = [int(v_i) for v_i in input().split()]
s = sum(v)

if n < 3:
    print(0)
elif s % 3 != 0:
    print(0)
else:
    print(souvenirs(int(s/3),n,v))
```

## Maximum value of an arithmetic expression

```python
#Parentheses
import math

def calcula(x, y, sim):
    if sim == '+':
        return x + y
    elif sim == '-':
        return x - y
    else:
        return x * y

def extremos(MaxValue, matriz_aux, i, j, simbolos):
    min_val = math.inf
    max_val = -math.inf
    for k in range(i, j):
        a = calcula(MaxValue[i][k], MaxValue[k+1][j], simbolos[k])
        b = calcula(MaxValue[i][k], matriz_aux[k+1][j], simbolos[k])
        c = calcula(matriz_aux[i][k], MaxValue[k+1][j], simbolos[k])
        d = calcula(matriz_aux[i][k], matriz_aux[k+1][j], simbolos[k])
        min_val = min(min_val, a, b, c, d)
        max_val = max(max_val, a, b, c, d)
    return min_val, max_val

def maximiza(numeros, simbolos):
    n = len(numeros)
    matriz_aux = [[None for x in range(n)] for x in range(n)]
    MaxValue = [[None for x in range(n)] for x in range(n)]

    for i in range(n):
        matriz_aux[i][i] = numeros[i]
        MaxValue[i][i] = numeros[i]

    for s in range(1, n):
        for i in range(0, n-s):
            j = i + s
            matriz_aux[i][j], MaxValue[i][j] = extremos(MaxValue, matriz_aux, i, j, simbolos)

    return MaxValue[0][n-1]

expresion = input()
simbolos = []
numeros = []

for i in expresion:
    if i in ['+', '-', '*']:
        simbolos.append(i)
    else:
        numeros.append(int(i))

print(maximiza(numeros, simbolos))
```
