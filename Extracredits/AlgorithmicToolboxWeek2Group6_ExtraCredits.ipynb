## **5. Fibonacci Number Again**

In this problem, your goal is to compute ùêπùëõ modulo ùëö, where ùëõ may be really huge: up to 1018. For such values of ùëõ, an algorithm looping for ùëõ iterations will not fit into one second for sure. Therefore we need to avoid such a loop.

To get an idea how to solve this problem without going through all ùêπùëñ for ùëñ from 0 to ùëõ, let‚Äôs see what happens when ùëö is small ‚Äî say, ùëö = 2 or ùëö = 3.

Pisano_periods.png

Take a detailed look at this table. Do you see? Both these sequences are periodic! For ùëö = 2, the period
is 011 and has length 3, while for ùëö = 3 the period is 01120221 and has length 8. Therefore, to compute,
say, ùêπ2015 mod 3 we just need to find the remainder of 2015 when divided by 8. Since 2015 = 251 ¬∑ 8 + 7, we
conclude that ùêπ2015 mod 3 = ùêπ7 mod 3 = 1.

This is true in general: for any integer ùëö ‚â• 2, the sequence ùêπùëõ mod ùëö is periodic. The period always
starts with 01 and is known as Pisano period.

---

###**Problem Description**

**Task.** Given two integers ùëõ and ùëö, output ùêπùëõ mod ùëö (that is, the remainder of ùêπùëõ when divided by ùëö).

**Input Format.** The input consists of two integers ùëõ and ùëö given on the same line (separated by a space).

**Constraints.** 1 ‚â§ ùëõ ‚â§ 1018, 2 ‚â§ ùëö ‚â§ 103.

**Output Format.** Output ùêπùëõ mod ùëö.




**Sample testcase**

```python

#ùêπ_239 mod 1000 = 39679027332006820581608740953902289877834488152161 (mod 1000) = 161.
#  --> Expected output: 161

calc_fib_huge(2816213588,239)
#  --> Expected output: 151
```

---

````python
def longitud_pisano(m):
    
    pisano1 = 0
    pisano2 = 1
    for i in range(m * m):
        pisano1, pisano2 = pisano2, (pisano1 + pisano2) % m
        
        if (pisano1 == 0 and pisano2 == 1):
            return i + 1

def calc_fib_huge(n, m):
    
    n = n % longitud_pisano(m) 

    if (n <= 1):
        return n
    
    fibonacci = []
    fibonacci.append(0)
    fibonacci.append(1)
    for i in range(2,n+1):
        fibonacci.append((fibonacci[i - 1] + fibonacci[i - 2]) % m) 
    return fibonacci[-1]
````

## **6. Last Digit of the Sum of Fibonacci Numbers**

Sum_Fibonacci.png

The goal in this problem is to find the last digit of a sum of the first ùëõ Fibonacci numbers.

---

###**Problem Description**

**Task.** Given an integer ùëõ, find the last digit of the sum $ùêπ_0 + ùêπ_1 + ¬∑ ¬∑ ¬∑ + ùêπ_ùëõ.$

**Input Format.** The input consists of a single integer ùëõ.

**Constraints.** 0 ‚â§ ùëõ ‚â§ 1014.

**Output Format.** Output the last digit of $ùêπ_0 + ùêπ_1 + ¬∑ ¬∑ ¬∑ + ùêπ_ùëõ.$


**Sample testcase**

```python
#run this sample testcase on a cell to help you validate your solution

last_digit_sum_fibonacci(3)
#ùêπ_0 + ùêπ_1 + ùêπ_2 + ùêπ_3 = 0 + 1 + 1 + 2 = 4.
#  --> Expected output: 4

last_digit_sum_fibonacci(100)
#ùêπ_0 + ... + ùêπ_100 = 927372692193078999175.
#  --> Expected output: 5
```

---

````python
def last_digit_sum_fibonacci(n):

    if (n <= 1):
        return n
    
    n+=2
    n = n % 60
    
    fibonacci = []
    fibonacci.append(0)
    fibonacci.append(1)
    for i in range(2,n+1):
        fibonacci.append((fibonacci[i - 1] + fibonacci[i - 2]) % 10)
    
    if(fibonacci[-1] == 0):
        return 9
    
    return fibonacci[-1]-1
````

## **7. Last Digit of the Sum of Fibonacci Numbers Again**

Fibonacci_rabbits.png

Now, we would like to find the last digit of a *partial* sum of Fibonacci numbers: ùêπùëö + ùêπùëö+1 + ¬∑ ¬∑ ¬∑ + ùêπùëõ.

---

###**Problem Description**

**Task.** Given two non-negative integers ùëö and ùëõ, where ùëö ‚â§ ùëõ, find the last digit of the sum $ùêπ_ùëö + ùêπ_{ùëö+1} + ¬∑ ¬∑ ¬∑ + ùêπ_ùëõ.$

**Input Format.** The input consists of two non-negative integers ùëö and ùëõ separated by a space.

**Constraints.** 0 ‚â§ ùëö ‚â§ ùëõ ‚â§ 1014.

**Output Format.** Output the last digit of $ùêπ_ùëö + ùêπ_{ùëö+1} + ¬∑ ¬∑ ¬∑ + ùêπ_ùëõ.$


**Sample testcase**

```python
#run this sample testcase on a cell to help you validate your solution

partial_sum_fibonacci(3,7)
#ùêπ_3 + ùêπ_4 + ùêπ_5 + ùêπ_6 + ùêπ_7 = 2 + 3 + 5 + 8 + 13 = 31.
#  --> Expected output: 1

partial_sum_fibonacci(10,10)
#ùêπ_10 = 55.
#  --> Expected output: 5

partial_sum_fibonacci(10,200)
#ùêπ_10 + ùêπ_11 + ... + ùêπ_200 =  734544867157818093234908902110449296423262.
#  --> Expected output: 2
```

---

````python
def partial_sum_fibonacci(m,n):

    if (n <= 1 and m <= 1):
        return n
    
    n+=2
    n = n % 60

    m+=1
    m = m % 60
    
    fibonacci = []
    fibonacci.append(0)
    fibonacci.append(1)

    if(n >= m):
      max = n

    else: 
      max = m


    for i in range(2,max+1):
        fibonacci.append((fibonacci[i - 1] + fibonacci[i - 2])%10)
    
    sum = fibonacci[n] - fibonacci[m] + 10
    sum %= 10
    
    return sum
````

## **8. Last Digit of the Sum of Squares of Fibonacci Numbers**

Since the brute force search algorithm for this problem is too slow (ùëõ may be as large as 1018), we need to come
up with a simple formula for $F^2_0 +F^2_1 + \dots +F^2_ùëõ$. 

The figure below represents the sum $F^2_1 +F^2_2 +F^2_3 +F^2_4 +F^2_5$ as the area of a rectangle with vertical side $F_5 = 5$ and horizontal side $F_5 + F_4 = 3 + 5 = F_6$.

Fib_squares.png

---

###**Problem Description**

**Task.** Compute the last digit of $F^2_0 + F^2_1 + ¬∑ ¬∑ ¬∑ + F^2_ùëõ.$

**Input Format.** Integer ùëõ.

**Constraints.** 0 ‚â§ ùëõ ‚â§ 1014.

**Output Format.** The last digit of $F^2_0 + F^2_1 + ¬∑ ¬∑ ¬∑ + F^2_ùëõ.$


**Sample testcase**

```python
#run this sample testcase on a cell to help you validate your solution

sum_fibonacci_squares(7)
#(ùêπ_0)^2 + (ùêπ_1)^2 + ... + (ùêπ_7)^2 = 0 + 1 + 1 + 4 + 9 + 25 + 64 + 169 = 273.
#  --> Expected output: 3

sum_fibonacci_squares(73)
#(ùêπ_0)^2 + ... + (ùêπ_73)^2 = 1052478208141359608061842155201.
#  --> Expected output: 1

sum_fibonacci_squares(1234567890)
#  --> Expected output: 0
```

---

````python
def sum_fibonacci_squares(n):

    if (n <= 1):
        return n

    n = n % 60
    
    fibonacci = []
    fibonacci.append(0)
    fibonacci.append(1)
    for i in range(2,n+2):
        fibonacci.append((fibonacci[i - 1] + fibonacci[i - 2])%10)
    
    return (fibonacci[n] * fibonacci[n+1])%10
````

````python
sum_fibonacci_squares(1234567890)
````
