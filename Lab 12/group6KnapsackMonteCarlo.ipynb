##Proporcin Exacta de las "Knapsack solutions" para el problema de la imagen.

image.png

```python
import numpy as np
from itertools import product

a = [1,1,2,4,12]
b = int(input('Ingrese la capacidad de la mochila\n'))
n = len(a)
perm = product([0,1], repeat=n)
k = 0
total = 0

for x in perm:
    if(np.dot(a,x)<=b):
        k += 1
    total += 1

p = k/total

print("La proporcin exacta de \"Knapsack Solutions\" es de", p)
print("El nmero exacto de \"Knapsack Solutions\" es de", k)
```

##Proporcin Estimada de "Knapsack solutions" utilizando Monte Carlo con 1000 y 10000 vectores aleatorios.

```python
import numpy as np

MonteCarlo = 1000

a = [1,2,3,4,5,6,7,9,10]
b = int(input('Ingrese la capacidad de la mochila\n'))
n = len(a)
k = 0
total = 0

for i in range(1, MonteCarlo + 1):
    x = list(np.random.randint(2, size=n))
    if(np.dot(a,x)<=b):
        k += 1
    total += 1

p = k/total

print("La proporcin estimada de \"Knapsack Solutions\" para", MonteCarlo, "vectores aleatorios es de", p)
print("El nmero estimado de \"Knapsack Solutions\" para", MonteCarlo, "vectores aleatorios es de", k)
```

##Proporcin Estimada de "Knapsack solutions" utilizando Monte Carlo con 10000, 100000 y 1000000 vectores aleatorios.

```python
import numpy as np

MonteCarlo = 10000

a = list(range(1,51))
bList = list(map(int,input().split()))
n = len(a)

for b in bList:
    k = 0
    total = 0
    for i in range(1, MonteCarlo + 1):
        x = list(np.random.randint(2, size=n))
        if(np.dot(a,x)<=b):
            k += 1
        total += 1

    p = k/total
    print("Capacidad =", b)
    print("La proporcin estimada de \"Knapsack Solutions\" para", MonteCarlo, "vectores aleatorios es de", p)
    print("El nmero estimado de \"Knapsack Solutions\" para", MonteCarlo, "vectores aleatorios es de", k, end="\n\n")
```
