## Dynamic Money Change

```python
import math

money = int(input())

MinNumCoins = [0]
for m in range(1,money+1):
    MinNumCoins.append(math.inf)
    for i in [1,3,4]:
        if m >= i:
            NumCoins = MinNumCoins[m-i] + 1
            if NumCoins < MinNumCoins[m]:
                MinNumCoins[m] = NumCoins

print(MinNumCoins[money])
```

##Primitive Calculator

```python
import math

n = int(input())

MinNumOperations = [0, 0]

for i in range(2, n+1):
    add1, x2, x3 = [math.inf]*3

    add1 = MinNumOperations[i-1] + 1 
    if i%2 == 0: x2 = MinNumOperations[int(i/2)] + 1
    if i%3 == 0: x3 = MinNumOperations[int(i/3)] + 1
    MinNumOperations.append(min(add1, x2, x3))

print(MinNumOperations[n])

IntermediateNumbers = [n]
while n > 1:
    if n%3 == 0 and MinNumOperations[n]-1 == MinNumOperations[int(n/3)]:
        IntermediateNumbers.append(int(n/3))
        n = int(n/3)
    elif n%2 == 0 and MinNumOperations[n]-1 == MinNumOperations[int(n/2)]:
        IntermediateNumbers.append(int(n/2))
        n = int(n/2)
    else:
        IntermediateNumbers.append(n-1)
        n -= 1

IntermediateNumbers = IntermediateNumbers[::-1]

for i in IntermediateNumbers[:-1]:
    print(i,end=" ")
print(IntermediateNumbers[-1])
```

## Edit Distance

```python
A = input()
B = input()

n = len(A)
m = len(B)

D = [[j for j in range(m+1)]]
D += [([i]+[0]*m) for i in range(1,n+1)]

for j in range(1,m+1):
    for i in range(1,n+1):
        insertion = D[i][j-1] + 1
        deletion = D[i-1][j] + 1
        match = D[i-1][j-1]
        mismatch = D[i-1][j-1] + 1
        if A[i-1] == B[j-1]:
            D[i][j] = min(insertion,deletion,match)
        else:
            D[i][j] = min(insertion,deletion,mismatch)

print(D[n][m])
```

##Longest common Subsequence of Two Sequences

```python
n = int(input())
a = [int(a_i) for a_i in input().split()]
m = int(input())
b = [int(b_j) for b_j in input().split()]

LCS2 = [[0]*(m+1) for fila in range(n+1)]

for i in range(1, n+1):
    for j in range(1, m+1):
        if a[i-1] == b[j-1]:
            LCS2[i][j] = LCS2[i-1][j-1] + 1
        if a[i-1] != b[j-1]:
            LCS2[i][j] = max(LCS2[i][j-1], LCS2[i-1][j])

print(int(LCS2[n][m]))
```

## Longest Common Subsequence of Three Sequences

```python
import numpy

n = int(input())
a = [int(a_i) for a_i in input().split()]
m = int(input())
b = [int(b_j) for b_j in input().split()]
l = int(input())
c = [int(c_k) for c_k in input().split()]

LCS3 = numpy.zeros((n+1, m+1, l+1))

for i in range(1, n+1):
    for j in range(1, m+1):
        for k in range(1, l+1):
            if a[i-1] == b[j-1] == c[k-1]:
                LCS3[i][j][k] = LCS3[i-1][j-1][k-1] + 1
            else:
                LCS3[i][j][k] = max(LCS3[i-1][j][k], LCS3[i][j-1][k], LCS3[i][j][k-1])

print(int(LCS3[n][m][l]))
```
